### Number

There are three numeric types in Python:
* int- Int, or integer, is a whole number, positive or negative, without decimals, of unlimited length.
* float - Float, or "floating point number" is a number, positive or negative, containing one or more decimals.
(Float can also be scientific numbers with an "e" to indicate the power of 10.)
* complex -Complex numbers are written with a "j" as the imaginary part:
Variables of numeric types are created when you assign a value to them:

```python
#int
x = 1
y = 35656222554887711
z = -3255522

print(type(x))
print(type(y))
print(type(z))
```

```python
#float
x = 3.143
y = 12E4
z = -87.7e100
print(y,z)
print(type(x))
print(type(y))
print(type(z))
```

```python
#Complex:
x = 3 + 5j  # 3 is the real part, 5j is the imaginary part
y = 5j      # purely imaginary number (real part is 0)
z = -5j     # purely imaginary number with negative sign

print(type(x),type(y),type(z))
print(x,y,z)
print(z.real)  # prints 3.0
print(z.imag)  # prints 5.0
print(x)       # Output: (3+5j), A complex number is not a tuple, even if it’s displayed with parentheses.
print(y)       # Output: 5j
print(z)       # Output: -5j
```

Parentheses around complex numbers in print output are simply to clarify the number’s grouping,

#remove duplicates
several ways to remove duplicates from a list in Python. The simplest way to remove duplicates is by converting a list to a set.
Using set()
We can use set() to remove duplicates from the list. However, this approach does not preserve the original order.


```python
a = [1, 2, 2, 3, 4, 4, 5, 'abc', 'abc']
# Remove duplicates by converting to a set
a = list(set(a))
print(a)
```

A set in Python is an unordered collection of unique elements—it does not allow duplicates.
When you convert a list to a set using set(your_list), all duplicate elements are automatically removed because sets cannot hold duplicates.
Then converting the set back to a list (list(set(your_list))) gives you a list with duplicates removed

### Set

A set can be used for storing a group of elements enclosed within {} 
set will not allow Duplicate Elements 
set will allow null values 
set will not maintain the insertion Order 
set is mutable

```python
#Creating a Set in Python:In Python, the most basic and efficient method for creating a set is using curly braces.
set1 = {1,12, 2, 2, 0,3.1, -3, 4}
print(set1)
```

```python
set1 = set()
print(set1)
```

```python
set1 = set(["monkey", "For", "zoo"])
print(set1)
```

```python
# Creating a Set with the use of a tuple
tup = ("monkey", "For", "zoo")
print(set(tup))
```

In set, the order of elements is not guaranteed to be the same as the order in which they were added. The output could vary each time we run the program. Also the duplicate items entered are removed by itself.
Sets do not support indexing. Trying to access an element by index (set[0]) raises a TypeError.

```python
#Adding Elements to a Set in Python
#We can add items to a set using add() method and update() method. add() method can be used to add only a single item. To add multiple items we use update() method.
# Creating a set
set1 = {1, 2, 3}
# Add one item
set1.add(0)
print(set1)
```

```python
# Add multiple items
set1.update([-5, 0.6])

print(set1)
```

```python
#Using remove() Method or discard() Method
#remove() method removes a specified element from the set. If the element is not present in the set, it raises a KeyError. discard() method also removes a specified element from the set. Unlike remove(), if the element is not found, it does not raise an error.
set1.discard(10)  # No error raised
print(set1)
```

```python
# Using Remove Method
set1.remove(3)
print(set1)  
set1.remove(13)
print(set1) 
```

```python
#clear() method removes all elements from the set, leaving it empty.
set1 = {1, 2, 3, 4, 5}
set1.clear()
print(set1)
```

```python
# Popping an element (removes a random item), takes no argument
set1 = {1, 2, 3, 4, 15}
set1.pop()
print(set1)
```

### Set Operations 

can be defined as the operations performed on two or more sets to obtain a single set containing a combination of elements from all the sets being operated upon. 

```python
#union
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union (elements in A OR B)
print(A | B)  # Output: {1, 2, 3, 4, 5, 6}
```

```python
# Intersection (elements in BOTH A AND B)
print(A & B)  
```

```python
# Difference (elements in A but NOT in B)
print(A - B) 

# Symmetric Difference (elements in A OR B, but NOT both)
print(A ^ B) 
```

### If ... Else

```python
If-Else is a fundamental conditional statement used for decision-making in programming.
If...Else statement allows to execution of specific blocks of code depending on the condition is True or False.
```

```python
age = 24
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

```python
score=71
if score >= 90:
    print("Grade: A")
elif score >= 70:
    print("Grade: B")
else:
    print("Grade: C")
```
